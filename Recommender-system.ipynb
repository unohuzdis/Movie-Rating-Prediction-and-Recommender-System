```python
import pandas as pd

r_cols = ['userId', 'movieId', 'rating']
df_ratings = pd.read_csv(
            'dataset/ratings.csv',
            usecols=r_cols,
            dtype={'userId': 'int32', 'movieId': 'int32', 'rating': 'float32'})

m_cols = ['movieId', 'title']
df_movies = pd.read_csv(
            'dataset/movies.csv',
            usecols=m_cols,
            dtype={'movieId': 'int32', 'title': 'str'})

df_merged = pd.merge(df_movies, df_ratings)

df_merged.head()
```

```python
userRatings = df_merged.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings.head()
```

```python
correlationMatrix = userRatings.corr()
correlationMatrix.head()
```

```python
correlationMatrix = userRatings.corr(method='pearson', min_periods=10)
correlationMatrix.head()
```

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print ("Adding sims for " + myRatings.index[i] + "...")
    # Retrieve similar movies to this one that I rated
    sims = correlationMatrix[myRatings.index[i]].dropna()
    # Scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Results based on similarity
print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```
